# Measure of central tendency
    MEAN   - The average of the dataset is described by a term called Mean.
    MEDIAN - The middle value of the sorted dataset is called Median, which is robust to outliers
    MODE   - The most frequent number in the dataset is called Mode.

# Formula for Mean:


$$  \sum_{k=1}^{n}\, \frac{Xi}{n}  $$ 

```python
print("Mean of Monthly income:",df['Mthly_HH_Income'].mean())
print("Mean of Monthly expenses:",df['Mthly_HH_Expense'].mean())
print("Mean of Annual Income:",df['Annual_HH_Income'].mean())
```

# Formula for Median:

$$  Y= \tilde{X} $$

```python
print("Median of Monthly income:",df['Mthly_HH_Income'].median())
print("Median of Monthly expenses:",df['Mthly_HH_Expense'].median())
print("Median of Annual Income:",df['Annual_HH_Income'].median())
```

# Implementation of Mode:

```python
print("Mode of Monthly income:",df['Mthly_HH_Income'].mode())
print("Mode of Monthly expenses:",df['Mthly_HH_Expense'].mode())
print("Mode of Annual Income:",df['Annual_HH_Income'].mode())
```

# Measure of Spread
    Variance           - Variance tells you the degree of spread of your dataset
    Standard Deviation - The standard deviation is a measure of the amount of variation or dispersion of a set of values.

# Formula for Variance:

$$ \sigma^2 = \frac{\sum_{i=1}^{n}(x_i - \mu)^2} {n} $$

```python
print("Variance of Monthly income:",round(df['Mthly_HH_Income'].var(),2))
print("Variance of Monthly expenses:",round(df['Mthly_HH_Expense'].var(),2))
print("Variance of Annual Income:",round(df['Annual_HH_Income'].var(),2))
```

# Formula for Standard Deviation:

$$ \sigma = \sqrt {\mu _2 } $$

```python
print("Standard Deviation of Monthly income:",round(df['Mthly_HH_Income'].std(),2))
print("Standard Deviation of Monthly expenses:",round(df['Mthly_HH_Expense'].std(),2))
print("Standard Deviation of Annual Income:",round(df['Annual_HH_Income'].std(),2))
```

# Skewness and Kurtosis
     Skewness - It measures symmetry of distribution. It is equal to 0 for a normal distribution.
     -> If skewness is less than -1 or greater than 1, the distribution is highly skewed.
     -> If skewness is between -1 and -0.5 or between 0.5 and 1, the distribution is moderately skewed.
     -> If skewness is between -0.5 and 0.5, the distribution is approximately symmetric.
     
     Kurtosis - It is the measure of peakedness. It tells you the height and sharpness of the central peak

# Measuring Skewness

```python
print("Skewness of Monthly income:",round(df['Mthly_HH_Income'].skew(),2))
print("Skewness of Monthly expenses:",round(df['Mthly_HH_Expense'].skew(),2))
print("Skewness of Annual Income:",round(df['Annual_HH_Income'].skew(),2))
```

# Measuring Kurtosis


```python
print("Kurtosis of Monthly income:",round(df['Mthly_HH_Income'].kurt(),2))
print("Kurtosis of Monthly expenses:",round(df['Mthly_HH_Expense'].kurt(),2))
print("Kurtosis of Annual Income:",round(df['Annual_HH_Income'].kurt(),2))
```

# Effect of mean, median and mode due to skewness
       -> If the distribution of data is skewed to the left, the mean is less than the median, which is often less than the mode. 
       -> If the distribution of data is skewed to the right, the mode is often less than the median, which is less than the mean.
       -> This is due to the presence of outliers in the data.
       -> Mean is always impacted by outliers, So that the data is skewed
       -> Whereas,In Median and Mode it cannot be affected by outliers. That is Median and mode are very robust to outliers.

# Q-Q Plot 
       -> A Q-Q plot is a scatterplot created by plotting two sets of quantiles against one another. 
       -> Q-Q plot is a way to test whether the data is normally distributed or not.
       -> When plotting the Q-Q plot, a 45 degree line is also plotted for reference.

# Implementation of Q-Q plot

```python
#Checking normality for the given data.
plt.figure(figsize=(12,8))
plt.subplot(231)
stats.probplot(df['Mthly_HH_Income'], dist="norm", plot=plt)
plt.title("Q-Q plot of Monthly income")
plt.subplot(232)
stats.probplot(df['Mthly_HH_Expense'], dist="norm", plot=plt)
plt.title("Q-Q plot of Monthly Expenses")
plt.subplot(233)
stats.probplot(df['Annual_HH_Income'], dist="norm", plot=plt)
plt.title("Q-Q plot of Annual income")
plt.show()
```

# Observation:
      -> It is observed that none of the data is normally distributed.
      -> All the data is skewed at a particular point.

# Box-Cox transformation
    -> A Box Cox transformation is a transformation of a non-normal dependent variables into a normal shape.
    -> Normality is an important assumption of many stastical techniques.
    -> The Box-Cox transformation transforms our data so that it closely resembles a normal distribution.

# Implementation of box-cox transformation

```python
#Box-Cox transformation of the given data
plt.figure(figsize=(12,8))

plt.subplot(231)
x_t,l = stats.boxcox(df['Mthly_HH_Income'])
sns.distplot(x_t)
plt.title("Normal plot of Monthly Income")

plt.subplot(232)
y_t,l = stats.boxcox(df['Mthly_HH_Expense'])
sns.distplot(y_t)
plt.title("Normal plot of Monthly Expenses")

plt.subplot(233)
z_t,l = stats.boxcox(df['Annual_HH_Income'])
sns.distplot(z_t)
plt.title("Normal plot of Annual Income")

plt.subplot(234)
stats.probplot(x_t, dist="norm", plot=plt)
plt.title("Q-Q plot of Monthly Income")

plt.subplot(235)
stats.probplot(y_t, dist="norm", plot=plt)
plt.title("Q-Q plot of Monthly Expense")

plt.subplot(236)
stats.probplot(z_t, dist="norm", plot=plt)
plt.title("Q-Q plot of Annual Income")
plt.show()
```

# Observations:
    -> After doing the Box cox transformation, it is observed that all the data are distributed normally.
    -> The Box-Cox power transformation is not a guarantee for normality. This is because it actually does not really check for normality.

# Features of Normal distribution
    -> The mean, mode and median are all equal. 
    -> The curve is symmetric at the center (i.e. around the mean, ). 
    -> Exactly half of the values are to the left of center and exactly half the values are to the right.
    -> Normally distributed data follows the 68-95-99.7 rule

# 68 - 95 - 99.7 Rule

```python
#Checking the feature for Monthly income data
mu_mthly_income = x_t.mean()
sigma_mthly_income = x_t.std()
one_std_right = mu_mthly_income + (1 * sigma_mthly_income)
one_std_left = mu_mthly_income - (1 * sigma_mthly_income)
two_std_right = mu_mthly_income + (2 * sigma_mthly_income)
two_std_left = mu_mthly_income - (2 * sigma_mthly_income)
three_std_right = mu_mthly_income + (3 * sigma_mthly_income)
three_std_left = mu_mthly_income - (3 * sigma_mthly_income)
```

```python
plt.figure(figsize=(20,10))
sns.set_style("darkgrid")
sns.distplot(x_t)

plt.axvline(mu_mthly_income, color='coral', label='Mean')

plt.axvline(one_std_right, color='yellow', label='Mean + 1SD')
plt.axvline(one_std_left, color='yellow', label='Mean - 1SD')
plt.axvline(two_std_right, color='green', label='Mean + 2SD')
plt.axvline(two_std_left, color='green', label='Mean - 2SD')
plt.axvline(three_std_right, color='blue', label='Mean + 3SD')
plt.axvline(three_std_left, color='blue', label='Mean - 3SD')
plt.legend();
```

```python
#Checking the feature for Monthly expense data
mu_mthly_expense = y_t.mean()
sigma_mthly_expense = y_t.std()
one_std_right = mu_mthly_expense + (1 * sigma_mthly_expense)
one_std_left = mu_mthly_expense - (1 * sigma_mthly_expense)
two_std_right = mu_mthly_expense + (2 * sigma_mthly_expense)
two_std_left = mu_mthly_expense - (2 * sigma_mthly_expense)
three_std_right = mu_mthly_expense + (3 * sigma_mthly_expense)
three_std_left = mu_mthly_expense - (3 * sigma_mthly_expense)
```

```python
plt.figure(figsize=(20,10))
sns.set_style("darkgrid")
sns.distplot(y_t)

plt.axvline(mu_mthly_expense, color='coral', label='Mean')

plt.axvline(one_std_right, color='yellow', label='Mean + 1SD')
plt.axvline(one_std_left, color='yellow', label='Mean - 1SD')
plt.axvline(two_std_right, color='green', label='Mean + 2SD')
plt.axvline(two_std_left, color='green', label='Mean - 2SD')
plt.axvline(three_std_right, color='blue', label='Mean + 3SD')
plt.axvline(three_std_left, color='blue', label='Mean - 3SD')
plt.legend();
```

```python
#Checking the feature for Annual income data
mu_annual_income = z_t.mean()
sigma_annual_income = z_t.std()
one_std_right = mu_annual_income + (1 * sigma_annual_income)
one_std_left = mu_annual_income - (1 * sigma_annual_income)
two_std_right = mu_annual_income + (2 * sigma_annual_income)
two_std_left = mu_annual_income - (2 * sigma_annual_income)
three_std_right = mu_annual_income + (3 * sigma_annual_income)
three_std_left = mu_annual_income - (3 * sigma_annual_income)
```

# Pareto distribution
    -> The Pareto distribution is a power-law probability distribution that is used in description of social, scientific, geophysical, actuarial, and many other types of observable phenomena.
    -> The Pareto distribution serves to show that the level of inputs and outputs is not always equal.
    -> It explains the famous 80-20 rule.
    

# Log normal distribution 
    -> Lognormal distribution plays an important role in probabilistic design because negative values of engineering phenomena are sometimes physically impossible. 
    -> Typical uses of lognormal distribution are found in descriptions of fatigue failure, failure rates, and other phenomena involving a large range of data.
    -> Both normal and lognormal distributions are used in statistical mathematics to describe the probability of an event occurring.
